```python
import pandas as pd
from nasspython import nass_api as na
# pip install nasspython: https://pypi.org/project/nasspython/
from statsmodels.tsa.arima_model import ARIMA
from pathlib import Path
import datetime as dt
```

```python
USDA_API_KEY = 'C6C2F629-A5BF-3F27-916A-4F15189AD588'
# function inputs
# NY/NJ/PA
state_list = ['NEW YORK','NEW JERSEY','PENNSYLVANIA']
start_yr = 2000
end_yr = 2020

# create list for years
year_list = [start_yr]
year = start_yr
while year < end_yr:
    year += 1
    year_list.append(year)
    
year_list = pd.to_datetime(year_list, format='%Y').year
year_list
```

```python
# Set up list to store USDA agricultural API data
vpa_list = []
year_list_new = []
state_list_new = []
# loop to pull data from API into dataframe then into empty lists
for yr in year_list:
    data = na.nass_data(USDA_API_KEY, year = yr, 
                    agg_level_desc = "STATE", 
                    short_desc = "AG LAND, CROPLAND - ASSET VALUE, MEASURED IN $ / ACRE"
                    )
    df = pd.DataFrame(data['data']).set_index("state_name")
    
    for state in state_list:
        vpa = df.loc[state,'Value']
        
        # clean strings into integers
        vpaint = ''
        for char in vpa:
            if char.isdigit() == True:
                vpaint = vpaint + char
        
        vpa_list.append(int(vpaint))
        year_list_new.append(yr)
        state_list_new.append(state)

print(f'{len(vpa_list)}, {len(year_list_new)}, {len(state_list_new)}')
```

```python
# create dictionary to convert into df
dict = {'Year': year_list_new, 'State': state_list_new, 'ValuePerAcre': vpa_list}
land_df = pd.DataFrame(dict)
pa_vpa = land_df[land_df['State'] == 'PENNSYLVANIA']['ValuePerAcre'].tolist()
ny_vpa = land_df[land_df['State'] == 'NEW YORK']['ValuePerAcre'].tolist()
nj_vpa = land_df[land_df['State'] == 'NEW JERSEY']['ValuePerAcre'].tolist()

# reorganize data into new df with states value per acre in diff cols
dict = {'Year': year_list, 'PAValuePerAcre': pa_vpa, 'NYValuePerAcre': ny_vpa, 'NJValuePerAcre': nj_vpa}
st_land_df = pd.DataFrame(dict)
st_land_df = st_land_df.set_index('Year')

st_land_df.plot(title='Historical PA/NY/NJ Farmland Value per Acre')
```

```python
# loop to find the 2021 predicted statewide value per acre based on historical data
# analysis uses an ARIMA model to forecast the value
predictions = {}
for col in st_land_df.columns:
    model = ARIMA(st_land_df[col], order=(0, 1, 1))
    results = model.fit()
    predict_df = pd.DataFrame(results.forecast(steps=2)[0])
    predictions[col] = predict_df.iloc[1,0]
predictions
```

```python
# pull in farmland properties that are currently for sale from csv
# filter for active listings
land_forsale_csv = Path("Ag_Land_ForSale.csv")
land_forsale_df = pd.read_csv(land_forsale_csv)
land_forsale_df = land_forsale_df[land_forsale_df['Sale Status'] == 'Active']

# land acreage converted to float from str
acitemlist = []
for item in land_forsale_df['Land Area (AC)']:
    acint = ''
    for char in item:
        if char.isdigit() == True:
            acint = acint + char
        elif char == '.':
            acint = acint + char
    acitemlist.append(float(acint))

land_forsale_df['Land Area (AC)'] = acitemlist

# df filtered to exclude small plots of land and blank values
land_forsale_df = land_forsale_df[land_forsale_df['Land Area (AC)'] >= 20]
land_forsale_df = land_forsale_df[land_forsale_df['Price Per AC Land'].notnull()]
```

```python
# remove commas from price per ac land strings
ppaitemlist = []
for item in land_forsale_df['Price Per AC Land']:
    ppaint = ''
    for char in str(item):
        if char.isdigit() == True:
            ppaint = ppaint + char
        elif char == '.':
            ppaint = ppaint + char
    ppaitemlist.append(ppaint)
    
land_forsale_df['Price Per AC Land'] = ppaitemlist

#convert str to float for price per ac land column
land_forsale_df['Price Per AC Land'] = pd.to_numeric(land_forsale_df['Price Per AC Land'])
```

```python
# add column with relevant state 2021 predicted values
state_avg = []
for item in land_forsale_df['State']:
    if item == 'PA':
        state_avg.append(predictions['PAValuePerAcre'])
    elif item == 'NY':
        state_avg.append(predictions['NYValuePerAcre'])
    elif item == 'NJ':
        state_avg.append(predictions['NJValuePerAcre'])
    else:
        state_avg.append(0)

land_forsale_df['state_avg'] = state_avg

# filter df for listings with price per acre's that are less than the state predicated value
potential_purchases_df = land_forsale_df[land_forsale_df['state_avg'] > land_forsale_df['Price Per AC Land']]
potential_purchases_df = potential_purchases_df[potential_purchases_df['Listing Broker Company'] != 'Research in Progress']
potential_purchases_df['difference'] = potential_purchases_df['Price Per AC Land'] - potential_purchases_df['state_avg']

potential_purchases_df.set_index('difference', inplace=True)

# create variables for output
options = potential_purchases_df['Address'].count()
cheapest = potential_purchases_df.index.min()
cheapest_address = potential_purchases_df.loc[cheapest, 'Address']
cheapest_city = potential_purchases_df.loc[cheapest, 'City']
cheapest_state = potential_purchases_df.loc[cheapest, 'State']
cheapest_contact = potential_purchases_df.loc[cheapest, 'Sale Company Contact']
contact_company = potential_purchases_df.loc[cheapest, 'Listing Broker Company']

potential_purchases_df.reset_index(inplace=True)

print(f'Model produced {options} potential options...') 
print(f'Call {cheapest_contact} at {contact_company} for {cheapest_address}, {cheapest_city} {cheapest_state}, the cheapeast option compared to the predicted 2021 statewide prices per acre.')
output_df = potential_purchases_df[['Address','City','County','Zip','State','Longitude','Latitude','Sale Price','Land Area (AC)',
                                    'Price Per AC Land','state_avg','Listing Broker Company','Sale Company Contact']]
output_df
```

```python
output_path = Path("output.csv")
output_df.to_csv(output_path)
```
